```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("backend/db.sqlite3")
```

```python
df = pd.read_parquet("nft_average_holding_period.parquet", engine="fastparquet")
df.to_sql("nft_avg_holding",con=con)
```

```python
df = pd.read_parquet("nft_diamond_hands.parquet", engine="fastparquet")
df.to_sql("nft_diamond_hands",con=con)
```

```python
df = pd.read_parquet("meved_entity.parquet")
df.to_sql("meved_entity",con=con)
```

```python
df = pd.read_parquet("phished_entity.parquet")
df
```

```python
df.to_sql("phished_entity",con=con)
```

```python
df = pd.read_parquet("address_db.parquet")
df.to_sql("address_labels",con=con)
```

```python
df = pd.read_parquet("mev_phish_stats.parquet")
df.columns = [ "_".join(list(tup)) for tup in df.columns.to_flat_index()]
df.to_sql("label_stats",con=con)
```
